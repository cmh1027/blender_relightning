```python
import cv2
import numpy as np

src = cv2.imread("chair/train/r_0.png")
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

orb = cv2.ORB_create(
    nfeatures=40000,
    scaleFactor=1.2,
    nlevels=8,
    edgeThreshold=31,
    firstLevel=0,
    WTA_K=2,
    scoreType=cv2.ORB_HARRIS_SCORE,
    patchSize=31,
    fastThreshold=20,
)

kp1, des1 = orb.detectAndCompute(gray, None)
print(kp1)
```
